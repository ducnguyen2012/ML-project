```python
#! crawl comment and overview
#! vào từng đường link một để lấy toàn bộ thông tin về phim đó. Mình sẽ chỉ lấy 3 cmt đầu nhiều reaction nhất của bộ phim đó. 
#! check thử url: 
#! comment đã oke. Giờ thì còn overview phim

#! data sẽ kiểu: tên phim + description + tên tác giả
cnt = 0
my_description = []
movie_name = []
for url in all_movie_link:
    # cnt += 1
#url = "https://myanimelist.net/anime/52991/Sousou_no_Frieren"
    page = requests.get(url)
    sleep(2)
    soup = BeautifulSoup(page.text,'html')
    
    path = soup.find_all(itemprop="description")

    sleep(2)
    for text in path:
    
        a = text.get_text().replace('\n','')
    
        my_description.append(a)
    path_name = soup.find_all(class_="title-name h1_bold_none")
    for name in path_name:
        temp = name.get_text().replace('\n','')
        movie_name.append(temp)
    # if cnt == 5:
    #     break
```
